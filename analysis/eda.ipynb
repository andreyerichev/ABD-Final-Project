## Описание датасета
Иcходный датасет содержат данные по месяцам жизни каждой компании-клиента интернет-эквайринга за период декабрь 2024 — февраль 2026.

Условные обозначения:
- `W` — размер окна скользящего среднего. Принимает значения `[2, 3, 6]`
- `Y` — дополнительные продукты `['p1', 'p2', 'p3']` и платёжный метод `'altpay_5'`
- `N` — платёжные методы `[1, 2, 3, 4, 5]`

### Категориальные переменные
#### Порядковые
- `month` — календарный месяц
- `first_trx_month_inn` — месяц, в котором у компании была первая боевая (т.е. не тестовая) транзакция. В интернет-эквайринге считается первым месяцем жизни клиента. Константный признак в рамках одной ИНН
- `last_active_month_inn` — последний месяц активности компании на момент текущего месяца `month`
- `first_month_Y` — месяц, в котором у ИНН впервые появился продукт `Y`. **Константный признак в рамках одной ИНН**
#### Номинальные
- `inn` — уникальный идентификатор компании. Константный признак в рамках одной ИНН
- `inn_status` — статус оттока компании на момент текущего месяца `month`. Принимает значения `['Active', 'New Churn', 'Churn', 'Reborn']`
- `top_mcc_group_inn` — группа MCC (merchant category code), по которой у данной компании был наибольший оборот в текущем месяце `month`
- `who_is_this_first_inn` — канал привлечения компании. Принимает значения: `['incoming', 'outcoming', 'referral']`. Константный признак в рамках одной ИНН
#### Бинарные
- `real_kam_on_inn` — наличие закреплённого Key Account Manager у компании на момент текущего месяца `month`
- `has_termination_start` — находится ли по данным ФНС компания в процессе ликвидации на момент текущего месяца `month`
- `has_recurrent_payments` — были ли рекуррентные платежи у компании на момент текущего месяца `month`
- `is_relevant_mcc_p3` — согласно существующему бизнес-правилу: подходят ли MCC, по которым у компании был оборот в текущем месяце `month`, для подключения продукта `P3`
- `is_relevant_mcc_altpay5` — согласно существующему бизнес-правилу: подходят ли MCC, по которым у компании был оборот в текущем месяце `month`, для подключения платёжного метода `altpay5`
- `is_relevant_turnover_altpay5` — согласно существующему бизнес-правилу: подходит ли оборот компании в текущем месяце `month` для подключения платёжного метода `altpay5`

### Количественные переменные
#### Непрерывные
- `cnt_trx` — кол-во транзакций компании в текущем месяце `month`. Признак логарифмирован и масштабирован
- `turnover` — оборот компании в текущем месяце `month`. Признак логарифмирован и масштабирован
- `altpayN_turnover` — оборот по платёжному методу `N` в текущем месяце `month`. Признак логарифмирован и масштабирован
- `avg_check` — средний чек компании в текущем месяце `month`. Признак логарифмирован и масштабирован
- `turnover_ma_Wm` — скользящий средний оборот компании с окном `W` на момент текущего месяца `month`. Признаки логарифмированы и масштабированы
- `turnover_ema_Wm` — экспоненциальный скользящий средний оборот компании с окном `W` на момент текущего месяца `month`. Признаки логарифмированы и масштабированы
- `cnt_trx_ma_Wm` — скользящее среднее кол-во транзакций компании с окном `W` на момент текущего месяца `month`. Признаки логарифмированы и масштабированы
- `cnt_trx_ema_Wm` — экспоненциальное скользящее среднее кол-во транзакций компании с окном `W` на момент текущего месяца `month`. Признаки логарифмированы и масштабированы

- `turnover_change_to_turnover_ema_Wm` — изменение оборота компании в текущем месяце `month` по отношению к экспоненциальному скользящему среднему обороту с окном `W`
- `cnt_trx_change_to_cnt_trx_ma_Wm` — изменение кол-ва транзакций компании в текущем месяце `month` по отношению к скользящему среднему кол-ву транзакций с окном `W`
- `cnt_trx_change_to_cnt_trx_ema_Wm` — изменение кол-ва транзакций компании в текущем месяце `month` по отношению к экспоненциальному скользящему среднему кол-ву транзакций с окном `W`
- `avg_check_change_to_avg_check_wma_Wm` — изменение кол-ва транзакций компании в текущем месяце `month` по отношению к скользящему среднему кол-ву транзакций с окном `W`

- `active_product_cnt` — кол-во активных продуктов на момент текущего месяца `month`. Не включает продукты, являющиеся таргетами

Среди них особенно выделяются первые месяцы, когда были подключены дополнительные продукты, доля строк с пропусками в этих столбцах составляет 91–99%, но т.к. это столбцы, из которых в дальнейшем будут составлены целевые переменные, обработку пропусков в них отложим. Начнём с категориального признака `top_mcc_group_inn`. Предположим, что он не принимает никакого значения, если компания на момент текущего месяца находится в оттоке. Нет транзакций <=> нет оборота => нет MCC => нет группы MCC. Проверим это предположение.

```python
base_df.query("top_mcc_group_inn.isna()").groupby('inn_status').size().sort_values(ascending=False)
```

Подавляющее большинство строк с пропусками в `top_mcc_group_inn` действительно имеют `inn_status == 'Churn'`, однако есть кейсы, когда компания является активной (т.е. имеет завершённые транзакции в отчётном месяце), но всё равно не имеет группу MCC. Такого физически не может быть в интернет-эквайринге, т.к. каждая транзакция имеет свой MCC. Возможно, это ошибка в существующей классификации групп MCC, и на самом деле MCC есть, но он не попал ни в какую группу, поэтому значение `NaN`. Заполним пропуски для оттёкших и активных клиентов разными значениями: для оттёкших заполним пропуски значением `'Churn'`, а для активных — `'No MCC Group'`.

```python
base_df.loc[
    (base_df['inn_status'].isin(['Churn', 'New Churn'])) & (base_df['top_mcc_group_inn'].isna()),
    'top_mcc_group_inn'
] = 'Churn'

base_df.loc[
    base_df['inn_status'].isin(['Active', 'Reborn']) & (base_df['top_mcc_group_inn'].isna()),
    'top_mcc_group_inn'
] = 'No MCC Group'
```

## Определение таргетов и разделение датасетов

В рамках данной работы таргетом будет являться факт подключения нового продукта у ИНН в следующем месяце. Соответственно, для каждого продукта необходимо собрать такой датасет, в котором по каждой ИНН будут только месяцы жизни строго до месяца подключения нового продукта. Продукты `'p1'` и `'p2'` по бизнес-правилу взаимоисключающие, поэтому в рамках данной работы для них будет собран один общий датасет. В нём таргетом будем считать по каждой ИНН строку с месяцем, предыдущим перед месяцем подключения первого продукта из множества `('p1', 'p2')`.

```python
def make_product_df(base_df: pd.DataFrame, product: str) -> pd.DataFrame:
    """Собирает датафрейм для переданного продукта и определяет таргет как последний месяц перед подключением"""
    df = base_df.copy()

    first_month_col = f'first_month_{product}'

    # Оставляем только строки до подключения продукта
    df = df[
        (df['month'] < df[first_month_col]) | (df[first_month_col].isna())
    ].reset_index(drop=True)

    df['target'] = (
        df['month'] + pd.DateOffset(months=1) == df[first_month_col]
    ).astype('int8')

    df = df.drop(columns=[first_month_col])

    # Устраняем Data Leakage по другим продуктам
    other_product_cols = [col for col in df.columns if col.startswith("first_month_")]
    for col in other_product_cols:
        df[col] = df[
            (df['month'] >= df[col]) | df[col].isna()
        ][col]

    return df


# Определяем общий таргет для продуктов p1 и p2
base_df['first_month_p1p2'] = base_df[['first_month_p1', 'first_month_p2']].min(axis=1)
base_df = base_df.drop(columns=['first_month_p1', 'first_month_p2'])

df_p1p2 = make_product_df(base_df=base_df, product='p1p2')
df_p3 = make_product_df(base_df=base_df, product='p3')
df_altpay5 = make_product_df(base_df=base_df, product='altpay5')

dfs = [df_p1p2, df_p3, df_altpay5]
```

Оценим распределение классов в датафреймах по каждому продукту.

```python
print("=== Баланс классов ===")
print("Всего наблюдений | Положительный класс | Доля положительного класса")
for df in dfs:
    share_of_positive = df.target.sum() / len(df)
    print(f"{len(df):<16} | {int(df.target.sum()):<19} | {share_of_positive:.2%}")
```

Наблюдаем экстремальный дисбаланс классов. Это ожидаемо, т.к. в исходном датасете собраны данные по месяцам жизни клиентов компании, привлечённых за всё время. Попробуем увеличить долю положительного класса за счёт удаления строк, где `inn_status in ('Churn', 'New Churn')`. Но сначала проверим, встречаются ли кейсы, когда клиент в статусе «Отток» по основному продукту (интернет-эквайринг) в следующем месяце подключил новый доп. продукт.

```python
print("=== Положительный класс ===")
print("Всего | В оттоке | Доля")
for df in dfs:
    positive_in_churn = df.query("inn_status in ('Churn', 'New Churn')").target.sum()
    share_of_positive_in_churn = positive_in_churn / df.target.sum()
    print(f"{df.target.sum():<5} | {int(positive_in_churn):<8} | {share_of_positive_in_churn:.2%}")
```

Такие кейсы встречаются, их доля составляет 20–40% от общего кол-ва положительных классов. В силу редкости объектов с положительным таргетом, их количественное уменьшение для нас крайне нежелательно. Однако, всё же проверим, как изменятся доли положительного класса в датасетах, если мы решим избавиться от строк с условием `inn_status in ('Churn', 'New Churn')`.

```python
print("=== Баланс классов (только активные ИНН) ===")
print("Всего наблюдений | Положительный класс | Доля положительного класса")
for df in dfs:
    df_active = df.query("inn_status not in ('Churn', 'New Churn')").copy()
    share_of_positive = df_active.target.sum() / len(df_active)
    print(f"{len(df_active):<16} | {int(df_active.target.sum()):<19} | {share_of_positive:.2%}")
```

Доля положительного класса увеличилась в 2–3 раза, однако значения по-прежнему экстремально малые. Примем это к сведению, экспортируем датасеты по продуктам с заполнеными пропусками и приступим непосредственно к EDA по каждому продукту.

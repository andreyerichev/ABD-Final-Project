# Обучение моделей P1P2

##### Загрузка данных

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
pd.set_option('display.precision', 4)

RANDOM_STATE = 1337
```

```python
from acquiring_xsell.data import load_data
from acquiring_xsell.preprocessing import process_dtypes, fill_missing_values
from acquiring_xsell.features import build_features, make_product_df

df = load_data()
df = process_dtypes(df)
df = fill_missing_values(df)
df = build_features(df)
df = make_product_df(df, "p1p2")

df.head()
```

## Случайный отбор

```python
from acquiring_xsell.preprocessing import train_test_month_split
from acquiring_xsell.metrics import format_metrics

train, test = train_test_month_split(df)
K_VALUES = [500, 1000, 1500, 2000]


monthly_ap_train = train.groupby('month')['target'].mean()
monthly_ap_test = test.groupby('month')['target'].mean()

monthly_precision_k_train = {}
monthly_precision_k_test = {}

monthly_recall_k_train = {}
monthly_recall_k_test = {}

monthly_expected_conversions_train = {}
monthly_expected_conversions_test = {}

top_k_train = {}
top_k_test = {}

for k in K_VALUES:

    monthly_precision_k_train[k] = train.groupby('month')['target'].mean()
    monthly_precision_k_test[k] = test.groupby('month')['target'].mean()

    monthly_recall_k_train[k] = k / train.groupby('month').size()
    monthly_recall_k_test[k] = k / test.groupby('month').size()

    monthly_expected_conversions_train[k] = train.groupby('month')['target'].mean() * k
    monthly_expected_conversions_test[k] = test.groupby('month')['target'].mean() * k

    top_k_train[k] = {
        'precision_k': monthly_precision_k_train[k].mean(),
        'recall_k': monthly_recall_k_train[k].mean(),
        'expected_conversions_k': monthly_expected_conversions_train[k].mean()
    }

    top_k_test[k] = {
        'precision_k': monthly_precision_k_test[k].mean(),
        'recall_k': monthly_recall_k_test[k].mean(),
        'expected_conversions_k': monthly_expected_conversions_test[k].mean()
    }


print("Random Selection")

print("\nTrain")
print(f"MAP: {monthly_ap_train.mean():.2%} ± {monthly_ap_train.std():.2%}")
display(format_metrics(pd.DataFrame(top_k_train).T))

print("\nTest")
print(f"MAP: {monthly_ap_test.mean():.2%} ± {monthly_ap_test.std():.2%}")
display(format_metrics(pd.DataFrame(top_k_test).T))
```

```python
from acquiring_xsell.preprocessing import train_test_month_split
from acquiring_xsell.metrics import monthly_average_precision_score, all_top_k_scores, format_metrics

train, test = train_test_month_split(df)
y_score_col = 'business_rule_score'

train[y_score_col] = np.where(train['rule_flag_p1p2'] == 1, train['turnover'], 0)
test[y_score_col] = np.where(test['rule_flag_p1p2'] == 1, test['turnover'], 0)


monthly_ap_train = monthly_average_precision_score(train, 'target', y_score_col)
monthly_ap_test = monthly_average_precision_score(test, 'target', y_score_col)

monthly_top_k_train = all_top_k_scores(train, 'target', y_score_col)
monthly_top_k_test = all_top_k_scores(test, 'target', y_score_col)


print("Business-Rule Selection")

print("\nTrain")
print(f"MAP: {monthly_ap_train.mean():.2%} ± {monthly_ap_train.std():.2%}")
display(format_metrics(monthly_top_k_train))

print("\nTest")
print(f"MAP: {monthly_ap_test.mean():.2%} ± {monthly_ap_test.std():.2%}")
display(format_metrics(monthly_top_k_test))
```

### Basic

Обучим логистическую регрессию на всех признаках. В рамках Pipeline применим One-Hot Encoding для категориальных признаков и StandardScaler для числовых.

#### Модель без регуляризации

```python
from acquiring_xsell.preprocessing import train_test_month_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from acquiring_xsell.metrics import monthly_average_precision_score


train, test = train_test_month_split(df)

X_train = train.drop(columns=['target', 'month'])
X_test = test.drop(columns=['target', 'month'])

y_test = test['target']
y_train = train['target']


num_cols = X_train.select_dtypes(include='number').columns
cat_cols = X_train.select_dtypes(include=['object', 'category']).columns

column_transformer = ColumnTransformer(transformers=[
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(drop='first', sparse_output=True, dtype=np.int8), cat_cols),
], remainder='passthrough')

lr_pipe = Pipeline([
    ('column_transformer', column_transformer),
    ('lr', LogisticRegression(random_state=RANDOM_STATE))
])

lr_pipe.fit(X_train, y_train)


train_eval = train.copy()
test_eval = test.copy()

train_eval['lr_proba'] = lr_pipe.predict_proba(X_train)[:, 1]
test_eval['lr_proba'] = lr_pipe.predict_proba(X_test)[:, 1]

monthly_ap_train = monthly_average_precision_score(train_eval, 'target', 'lr_proba')
monthly_ap_test = monthly_average_precision_score(test_eval, 'target', 'lr_proba')

print("Logistic Regression")
print(f"Train MAP: {monthly_ap_train.mean():.2%} ± {monthly_ap_train.std():.2%}")
print(f"Test  MAP: {monthly_ap_test.mean():.2%} ± {monthly_ap_test.std():.2%}")
```

### VIF

#### Отбор признаков

Попробуем итеративно убирать признак с самым большим VIF, пока наибольшее значение VIF в датасете не станет превышать 10.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from acquiring_xsell.preprocessing import train_test_month_split


def get_high_vif_features(data: pd.DataFrame) -> list:

    df = data.copy()
    df_numeric = df.select_dtypes(include='number')

    high_vif_features = []
    max_vif_value = np.inf

    print("High VIF Features\n")

    while max_vif_value > 10:
        curr_df = df_numeric.drop(columns=high_vif_features)

        vif = pd.Series(
            index=curr_df.columns.tolist(),
            data=[variance_inflation_factor(curr_df, i) for i in range(len(curr_df.columns))],
        )

        max_vif_feature = vif.idxmax()
        max_vif_value = vif.max()

        if max_vif_value < 10:
            break
        
        high_vif_features.append(max_vif_feature)

        print(f"{max_vif_feature}, VIF={max_vif_value:.2f}")

    return high_vif_features


train, test = train_test_month_split(df)

HIGH_VIF_FEATURES = get_high_vif_features(train)
```

```python
from acquiring_xsell.preprocessing import train_test_month_split
from acquiring_xsell.model_selection import MonthSeriesSplit
from sklearn.model_selection import ParameterSampler
from lightgbm import LGBMClassifier, early_stopping
from acquiring_xsell.metrics import monthly_average_precision_score


train, test = train_test_month_split(df)

X_train = train.drop(columns=['target', 'month'])
X_test = test.drop(columns=['target', 'month'])

y_train = train['target']
y_test = test['target']

cv = MonthSeriesSplit()


param_grid = {
    "learning_rate": [0.01, 0.03, 0.05],
    "max_depth": [3, 5, 7],
    "num_leaves": [15, 31, 63],
    "min_child_samples": [50, 100, 200],
    "subsample": [0.7, 0.8, 0.9],
    "colsample_bytree": [0.7, 0.8, 1.0],
}

best_score = 0
best_params = None

for params in ParameterSampler(param_grid, n_iter=30, random_state=RANDOM_STATE):
    
    fold_scores = []
    for train_idx, valid_idx in cv.split(train):

        X_train_cv = X_train.loc[train_idx]
        y_train_cv = y_train.loc[train_idx]

        X_valid_cv = X_train.loc[valid_idx]
        y_valid_cv = y_train.loc[valid_idx]

        model = LGBMClassifier(
            **params,
            random_state=RANDOM_STATE,
            n_estimators=5000,
            n_jobs=-1,
            verbose=-1
        )

        model.fit(
            X_train_cv, y_train_cv,
            eval_set=[(X_valid_cv, y_valid_cv)],
            eval_metric="average_precision",
            callbacks=[early_stopping(50)],
        )

        valid_eval = train.loc[valid_idx].copy()
        valid_eval["score"] = model.predict_proba(X_valid_cv)[:, 1]

        fold_score = monthly_average_precision_score(valid_eval, 'target', 'score').mean()
        fold_scores.append(fold_score)

    mean_score = np.mean(fold_scores)
    std_score = np.std(fold_scores)

    if mean_score > best_score:
        best_score = mean_score
        best_std = std_score
        best_params = params
        print(f"New Best CV MAP: {mean_score:.2%} ± {std_score:.2%}")


train, valid, test = train_test_month_split(df, get_valid=True)

X_train = train.drop(columns=['target', 'month'])
X_valid = valid.drop(columns=['target', 'month'])
X_test = test.drop(columns=['target', 'month'])

y_train = train['target']
y_valid = valid['target']
y_test = test['target']


lgbm_conservative = LGBMClassifier(
    **best_params,
    random_state=RANDOM_STATE,
    n_estimators=5000,
    n_jobs=-1,
    verbose=-1
)

lgbm_conservative.fit(
    X_train,
    y_train,
    eval_set=[(X_valid, y_valid)],
    eval_metric="average_precision",
    callbacks=[early_stopping(50)]
)


train_eval = train.copy()
valid_eval = valid.copy()
test_eval = test.copy()

train_eval["score"] = lgbm_conservative.predict_proba(X_train)[:, 1]
valid_eval["score"] = lgbm_conservative.predict_proba(X_valid)[:, 1]
test_eval["score"] = lgbm_conservative.predict_proba(X_test)[:, 1]

train_metrics = monthly_average_precision_score(train_eval, 'target', 'score')
valid_metrics = monthly_average_precision_score(valid_eval, 'target', 'score')
test_metrics = monthly_average_precision_score(test_eval, 'target', 'score')


print("\nLightGBM Conservative")
print(f"Best Iteration: {lgbm_conservative.best_iteration_}")
print(f"Best Params: {best_params}")
print(f"Train MAP: {train_metrics.mean():.2%} ± {train_metrics.std():.2%}")
print(f"Valid MAP: {valid_metrics.mean():.2%} ± {valid_metrics.std():.2%}")
print(f"Test  MAP: {test_metrics.mean():.2%} ± {test_metrics.std():.2%}")
print(f"CV    MAP: {best_score:.2%} ± {best_std:.2%}")
```

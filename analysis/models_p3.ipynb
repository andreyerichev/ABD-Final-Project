## Метрики качества
Основной метрикой качества моделей является Precision@K, поскольку бизнес-задача заключается в эффективном распределении ограниченного ресурса отдела продаж. В рамках задачи capacity отдела составляет 2000 предложений в месяц, что определяет основное значение K. Дополнительно рассматриваются значения K=500, K=1000 и K=1500 для анализа устойчивости качества ранжирования.

Для оценки относительного качества моделей используется метрика Uplift — отношение Precision@K рассматриваемой модели к Precision@K базовых подходов. В качестве базовых подходов используются:

1. Случайный отбор клиентов
2. Отбор на основе текущих бизнес-правил

Дополнительно для оценки качества ранжирования анализируется зависимость Precision@K от K, что позволяет оценить способность модели выделять наиболее релевантных клиентов в верхней части ранжированного списка.

Также используется метрика Recall@K, отражающая долю всех целевых событий (подключений продукта), попадающих в top-K клиентов, что позволяет оценить полноту охвата моделью потенциально релевантных клиентов.

Для интерпретации результатов в бизнес-терминах рассчитывается ожидаемое количество подключений (Expected Conversions), определяемое как произведение Precision@K на K.

## Logistic Regression

Обучим логистическую регрессию на всех признаках. В рамках Pipeline применим One-Hot Encoding для категориальных признаков, масштабирование числовых признаков через StandardScaler и L1-регуляризацию, чтобы модель с её помощью могла самостоятельно сделать feature selection.

```python
from acquiring_xsell.preprocessing import train_test_month_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from acquiring_xsell.metrics import calculate_metrics


train, test = train_test_month_split(df, border_month='2025-09-01')

X_train = train.drop(columns=['target', 'month'])
y_train = train['target']

X_test = test.drop(columns=['target', 'month'])
y_test = test['target']


num_cols = X_train.select_dtypes(include='number').columns
cat_cols = X_train.select_dtypes(include=['object', 'category']).columns

column_transformer = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(drop='first', sparse_output=True, dtype=np.int8), cat_cols),
    ],
    remainder='passthrough'
)


c_values = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10]
lr_models = {}

print(f"Logistic Regression p3")

for c in c_values:
    lr_pipe = Pipeline([
        ('column_transformer', column_transformer),
        ('lr', LogisticRegression(
                        max_iter=100_000,
                        solver="liblinear",
                        l1_ratio=1,
                        C=c,
                        random_state=RANDOM_STATE
                    ))
    ])

    lr_pipe.fit(X_train, y_train)

    lr_models[c] = lr_pipe

    train_eval = train.copy()
    train_eval['lr_proba'] = lr_pipe.predict_proba(X_train)[:, 1]
    
    test_eval = test.copy()
    test_eval['lr_proba'] = lr_pipe.predict_proba(X_test)[:, 1]

    print("\n", "="*30, f"C={c}", "="*30)

    print(f"\nTrain")
    calculate_metrics(data=train_eval, y_score='lr_proba')

    print(f"\nTest")
    calculate_metrics(data=test_eval, y_score='lr_proba')
```

```python
def print_lr_coefficients(lr_model):
    feature_names = []

    num_cols = lr_model.named_steps['column_transformer'].transformers_[0][2]
    cat_encoder = lr_model.named_steps['column_transformer'].transformers_[1][1]
    cat_cols = lr_model.named_steps['column_transformer'].transformers_[1][2]

    feature_names.extend(num_cols)
    cat_feature_names = cat_encoder.get_feature_names_out(cat_cols)
    feature_names.extend(cat_feature_names)

    coefs = lr_model.named_steps['lr'].coef_[0]
    coef_df = pd.DataFrame({'feature': feature_names, 'coef': coefs})

    top_10 = coef_df.sort_values('coef', ascending=False).head(10)
    bottom_10 = coef_df.sort_values('coef', ascending=True).head(10)
    zero_coefs = coef_df[coef_df['coef'] == 0]

    print("Топ-10 признаков с наибольшими coef_:")
    print(top_10.to_string(index=False))

    print("\nТоп-10 признаков с наименьшими coef_:")
    print(bottom_10.to_string(index=False))

    zero_coef_share = len(zero_coefs) / len(coef_df)
    print(f"\nДоля признаков с коэффициентом 0: {zero_coef_share:.2%}")

    print("\nПризнаки с коэффициентом 0:")
    if zero_coefs.empty:
        print("Нет признаков с коэффициентом 0.")
    else:
        print(zero_coefs.to_string(index=False))


print_lr_coefficients(lr_models[0.05])
```

В процессе анонимизации признаки `['turnover', 'cnt_trx', 'avg_check']` и их производные были масштабированы с использованием глобального `MinMaxScaler`. Это приводит к снижению сопоставимости значений во времени, к тому же практически все из них были убраны в ноль L-1 регуляризацией, поэтому также для сравнения обучим логистическую регрессию без данных признаков.

```python
from acquiring_xsell.preprocessing import train_test_month_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from acquiring_xsell.metrics import calculate_metrics

PRE_SCALED_FEATURES = [
    'cnt_trx',
    'turnover',
    'avg_check',
    'turnover_ma_2m',
    'turnover_ema_2m',
    'turnover_ma_3m',
    'turnover_ema_3m',
    'turnover_ma_6m',
    'turnover_ema_6m',
    'cnt_trx_ma_2m',
    'cnt_trx_ema_2m',
    'cnt_trx_ma_3m',
    'cnt_trx_ema_3m',
    'cnt_trx_ma_6m',
    'cnt_trx_ema_6m',
    'avg_check_wma_2m',
    'avg_check_wma_3m',
    'avg_check_wma_6m',
    'altpay1_turnover',
    'altpay2_turnover',
    'altpay3_turnover',
    'altpay4_turnover',
    'altpay5_turnover',
    'altpay6_turnover',
]


train, test = train_test_month_split(df, border_month='2025-09-01')

X_train = train.drop(columns=['target', 'month'] + PRE_SCALED_FEATURES)
y_train = train['target']

X_test = test.drop(columns=['target', 'month'] + PRE_SCALED_FEATURES)
y_test = test['target']


num_cols = X_train.select_dtypes(include='number').columns
cat_cols = X_train.select_dtypes(include=['object', 'category']).columns

column_transformer = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(drop='first', sparse_output=True, dtype=np.int8), cat_cols),
    ],
    remainder='passthrough'
)


c_values = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10]
lr_models_wo_scaled_featuers = {}

print(f"Logistic Regression p3 without scaled features, y_score='lr_proba'")

for c in c_values:
    lr_pipe = Pipeline([
        ('column_transformer', column_transformer),
        ('lr', LogisticRegression(
                        max_iter=100_000,
                        solver="liblinear",
                        l1_ratio=1,
                        C=c,
                        random_state=RANDOM_STATE
                    ))
    ])

    lr_pipe.fit(X_train, y_train)

    lr_models_wo_scaled_featuers[c] = lr_pipe

    train_eval = train.copy()
    train_eval['lr_proba'] = lr_pipe.predict_proba(X_train)[:, 1]
    
    test_eval = test.copy()
    test_eval['lr_proba'] = lr_pipe.predict_proba(X_test)[:, 1]

    print("\n", "="*30, f"C={c}", "="*30)

    print(f"\nTrain")
    calculate_metrics(data=train_eval, y_score='lr_proba')

    print(f"\nTest")
    calculate_metrics(data=test_eval, y_score='lr_proba')
```

Попробуем подобрать гиперпараметры, чтобы улучшить качество.

```python
from acquiring_xsell.preprocessing import train_test_month_split
from sklearn.model_selection import ParameterSampler
from lightgbm import LGBMClassifier, early_stopping
from acquiring_xsell.metrics import calculate_metrics, precision_k


train, valid, test = train_test_month_split(df, get_valid=True)

X_train = train.drop(columns=['target', 'month'])
y_train = train['target']

X_valid = valid.drop(columns=['target', 'month'])
y_valid = valid['target']

X_test = test.drop(columns=['target', 'month'])
y_test = test['target']


# pos_weight = (1 - 0.00206) / 0.00206 ≈ 486
# Перебираем вокруг этого значения + недо/перевес

param_grid = {
    "scale_pos_weight": [100, 250, 486, 700],
    "num_leaves": [31, 63, 127],
    "min_child_samples": [50, 100, 200],
    "subsample": [0.7, 0.9],
    "subsample_freq": [1],
    "colsample_bytree": [0.7, 0.9],
    "reg_lambda": [0.1, 1.0, 5.0],
    "learning_rate": [0.05],
    "n_estimators": [1000],
}

best_score = 0
best_params = None

for params in ParameterSampler(param_grid, n_iter=30, random_state=RANDOM_STATE):
    
    lgbm = LGBMClassifier(**params, objective="binary", random_state=RANDOM_STATE)
    lgbm.fit(
        X_train, y_train,
        eval_set=[(X_valid, y_valid)],
        eval_metric="auc",  
        callbacks=[early_stopping(50)],
    )

    valid_tmp = valid.copy()
    valid_tmp['lgbm_score'] = lgbm.predict_proba(X_valid)[:, 1]

    score = precision_k(valid_tmp, y_score_col='lgbm_score', k=2000)
    
    if score > best_score:
        best_score = score
        best_params = params
        best_lgbm = lgbm
    print(f"Current Best Precision@2000: {best_score:.2%}")


valid_tmp = valid.copy()
valid_tmp['lgbm_score'] = best_lgbm.predict_proba(X_valid)[:, 1]

test_tmp = test.copy()
test_tmp['lgbm_score'] = best_lgbm.predict_proba(X_test)[:, 1]

print("\nLightGBM p3")

print("\nValid")
calculate_metrics(
    valid_tmp,
    y_score="lgbm_score"
)

print("\nTest")
calculate_metrics(
    test_tmp,
    y_score="lgbm_score"
)
```

```python
from acquiring_xsell.preprocessing import train_test_month_split
from sklearn.model_selection import ParameterSampler
from catboost import CatBoostClassifier
from acquiring_xsell.metrics import calculate_metrics, precision_k


train, valid, test = train_test_month_split(df, get_valid=True)

X_train = train.drop(columns=['target', 'month'])
y_train = train['target']

X_valid = valid.drop(columns=['target', 'month'])
y_valid = valid['target']

X_test = test.drop(columns=['target', 'month'])
y_test = test['target']

cat_cols = X_train.select_dtypes(include=['object', 'category']).columns.tolist()

param_grid = {
    "scale_pos_weight": [100, 250, 486, 700],
    "depth": [4, 6, 8],
    "min_data_in_leaf": [50, 100, 200],
    "subsample": [0.7, 0.9],
    "colsample_bylevel":[0.7, 0.9],
    "l2_leaf_reg": [1, 3, 10, 30],
    "iterations": [1000],
    "learning_rate": [0.05],
}

best_score = 0
best_params = None
best_cb = None

for params in ParameterSampler(param_grid, n_iter=30, random_state=RANDOM_STATE):

    cb = CatBoostClassifier(
        **params,
        loss_function="Logloss",
        cat_features=cat_cols,
        random_seed=RANDOM_STATE,
        early_stopping_rounds=50,
        eval_metric="AUC",
        verbose=0,
    )

    cb.fit(
        X_train, y_train,
        eval_set=(X_valid, y_valid),
    )

    valid_tmp = valid.copy()
    valid_tmp['cb_score'] = cb.predict_proba(X_valid)[:, 1]

    score = precision_k(valid_tmp, y_score_col='cb_score', k=2000)

    if score > best_score:
        best_score = score
        best_params = params
        best_cb = cb

    print(f"Current Best Precision@2000: {best_score:.2%}")

print("\nBest params:", best_params)
print("Best iteration:", best_cb.get_best_iteration())

valid_tmp = valid.copy()
test_tmp = test.copy()

valid_tmp['cb_score'] = best_cb.predict_proba(X_valid)[:, 1]
test_tmp['cb_score']  = best_cb.predict_proba(X_test)[:, 1]

print(f"CatBoost p3")

print("\nValid")
calculate_metrics(valid_tmp, y_score="cb_score")

print("\nTest")
calculate_metrics(test_tmp, y_score="cb_score")
```
